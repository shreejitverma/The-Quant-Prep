### Getting data from yahoo finance

```python
import pandas as pd
import numpy as np
import pandas_datareader.data as web
from datetime import datetime
```

#### To get the data of a single stock from yahoo

```python
start = datetime(2010,1,1)
end = datetime(2020,12,31)

# let's get data for a single stock, for eg. RELIANCE
# On yahoo the ticker symbol is "RELIANCE.NS"

df = web.DataReader("RELIANCE.NS", 'yahoo', start, end)
df.head()
```

#### To get the data of multiple stocks from yahoo at a single time

```python
stocks = ['^NSEI','^NSEBANK','RELIANCE.NS', 'SBIN.NS'] #make sure the name matches the ticker symbol on the yahoo finance site
start = datetime(2010,1,1)
end = datetime(2020,12,31)

df = web.DataReader(stocks, 'yahoo', start, end)
df.head()
```
